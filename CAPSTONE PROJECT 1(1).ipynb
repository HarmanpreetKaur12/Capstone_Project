```python
#removing outliers(Z-score treatment)
# In this technique, we cap our outliers data and make the limit i.e, above a particular value or less than that value, 
#all the values will be considered as outliers, and the number of outliers in the dataset gives that capping number.

#For Normal distributions: Use empirical relations of Normal distribution.
#The data points which fall below mean-3*(sigma) or above mean+3*(sigma) are outliers.
#where mean and sigma are the average value and standard deviation of a particular column.

#Finding the Boundary Values

print("Highest allowed of age ",data['age'].mean() + 3*data['age'].std())
print("Lowest allowed of age",data['age'].mean() - 3*data['age'].std())

print("Highest allowed of fnlwgt",data['fnlwgt'].mean() + 3*data['fnlwgt'].std())
print("Lowest allowed of fnlwgt",data['fnlwgt'].mean() - 3*data['fnlwgt'].std())

print("Highest allowed of educational-num",data['educational-num'].mean() + 3*data['educational-num'].std())
print("Lowest allowed of educational-num",data['educational-num'].mean() - 3*data['educational-num'].std())

#print("Highest allowed",data['gender'].mean() + 3*data['gender'].std())
#print("Lowest allowed",data['gender'].mean() - 3*data['gender'].std())

print("Highest allowed of capital-gain",data['capital-gain'].mean() + 3*data['capital-gain'].std())
print("Lowest allowed of capital-gain",data['capital-gain'].mean() - 3*data['capital-gain'].std())

print("Highest allowed of capital-loss",data['capital-loss'].mean() + 3*data['capital-loss'].std())
print("Lowest allowed of capital-loss",data['capital-loss'].mean() - 3*data['capital-loss'].std())

print("Highest allowed of hours-per-week",data['hours-per-week'].mean() + 3*data['hours-per-week'].std())
print("Lowest allowed of hours-per-week",data['hours-per-week'].mean() - 3*data['hours-per-week'].std())

#print("Highest allowed of income",data['income'].mean() + 3*data['income'].std())
#print("Lowest allowed of income",data['income'].mean() - 3*data['income'].std())
```

```python
 #  Finding the Outliers
 #(Bitwise  Or (!) The result in each position is 0 if both bits are 0, while otherwise the result is 1.)

 data[(data['age'] > 79.28838133208404) |(data['age'] < -1.583990561831719)] 
 data[(data['fnlwgt'] > 506809.5920015615) | (data['fnlwgt'] < -127177.79834286292)]
 data[(data['educational-num'] > 17.788044955224105) | (data['educational-num'] <2.451870383421312)]
 data[(data['capital-gain'] > 8195.875932468383) | (data['capital-gain'] < -7017.486386452513)]
 data[(data['capital-loss'] > 1300.3210418401006) | (data['capital-loss'] < -1124.2070810167277)]
 data[(data['hours-per-week'] > 75.96451214978259) | (data['hours-per-week'] < 5.034575732884335)]
```

```python
#Trimming of Outliers
#(Bitwise And (&) if both bits in the compared position are 1, the bit in the resulting binary representation is 1 
 #(1 Ã— 1 = 1); otherwise, the result is 0)
    
data = data[(data['age'] < 79.28838133208404) & (data['age'] > -1.583990561831719)]
data = data[(data['fnlwgt'] < 506809.5920015615) & (data['fnlwgt'] > -127177.79834286292)]
data = data[(data['educational-num'] < 17.788044955224105) & (data['educational-num'] >2.451870383421312)]
data = data[(data['capital-gain'] <8195.875932468383) & (data['capital-gain'] > -7017.486386452513)]
data = data[(data['capital-loss'] < 1300.3210418401006) & (data['capital-loss'] > -1124.2070810167277)]
data = data[(data['hours-per-week'] < 75.96451214978259) & (data['hours-per-week'] > 5.034575732884335)]
```

```python
data
```

```python
#Capping on Outliers

upper_limit = data['age'].mean() + 3*data['age'].std()
lower_limit = data['age'].mean() - 3*data['age'].std()
upper_limit = data['fnlwgt'].mean() + 3*data['fnlwgt'].std()
lower_limit = data['fnlwgt'].mean() - 3*data['fnlwgt'].std()
upper_limit = data['educational-num'].mean() + 3*data['educational-num'].std()
lower_limit = data['educational-num'].mean() - 3*data['educational-num'].std()
upper_limit = data['capital-gain'].mean() + 3*data['capital-gain'].std()
lower_limit = data['capital-gain'].mean() - 3*data['capital-gain'].std()
upper_limit = data['capital-loss'].mean() + 3*data['capital-loss'].std()
lower_limit = data['capital-loss'].mean() - 3*data['capital-loss'].std()
upper_limit = data['hours-per-week'].mean() + 3*data['hours-per-week'].std()
lower_limit = data['hours-per-week'].mean() - 3*data['hours-per-week'].std()
```

```python
#Capping on Outliers

data['age'] = np.where(
    data['age']>upper_limit,
    upper_limit,
    np.where(
        data['age']<lower_limit,
        lower_limit,
        data['age']
    )
)

data['fnlwgt'] = np.where(
    data['fnlwgt']>upper_limit,
    upper_limit,
    np.where(
        data['fnlwgt']<lower_limit,
        lower_limit,
        data['fnlwgt']
    )
)

data['educational-num'] = np.where(
    data['educational-num']>upper_limit,
    upper_limit,
    np.where(
        data['educational-num']<lower_limit,
        lower_limit,
        data['educational-num']
    )
)

data['capital-gain'] = np.where(
    data['capital-gain']>upper_limit,
    upper_limit,
    np.where(
        data['capital-gain']<lower_limit,
        lower_limit,
        data['capital-gain']
    )
)

data['capital-loss'] = np.where(
    data['capital-loss']>upper_limit,
    upper_limit,
    np.where(
        data['capital-loss']<lower_limit,
        lower_limit,
        data['capital-loss']
    )
)

data['hours-per-week'] = np.where(
    data['hours-per-week']>upper_limit,
    upper_limit,
    np.where(
        data['hours-per-week']<lower_limit,
        lower_limit,
        data['hours-per-week']
    )
)
```

# K Fold Cross Validation

```python
#Because it ensures that every observation from the original dataset has the chance of appearing in training and test set.
#Cross validation (CV) is one of the technique used to test the effectiveness of a machine learning models, 
#it is also a re-sampling procedure used to evaluate a model if we have a limited data. To perform CV we need to keep aside a 
#sample/portion of the data on which is not used to train the model, later use this sample for testing/validating.
 
#it generally results in a less biased model compare to other methods. Because it ensures that every observation from the
#original dataset has the chance of appearing in training and test set. This is one among the best approach if we have a 
#limited input data. 

from sklearn.model_selection import KFold
kfold = KFold(n_splits=5)   #Split the entire data randomly into K folds value should be between 5 to 10
```

```python
# Modeling step Test differents algorithm
classifiers1 = []     #to recognize objects and being able to separate them into categories.
#it helps us segregate vast 
                       #quantities of data into discrete values,  
```

```python
# ? extracts models describing data classes
# A classification model reads some input and generates an output that classifies the input into 
#some category.
classifiers1.append(KNeighborsClassifier())   #the KNeighborsClassifier looks for the 5 nearest neighbors.
classifiers1.append(LogisticRegression())
classifiers1.append(GaussianNB())  #helps in building the fast machine learning models that can make quick predictions.
classifiers1.append(RandomForestClassifier())# It provides higher accuracy through cross validation. 
#Random forest classifier will handle the missing values and maintain the accuracy of a large 
#proportion of data.
```

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracy_results1 = []
for a in classifiers1:
    accuracy_results1.append(cross_val_score(a, train_input_SMOTE, train_output_SMOTE, scoring= "accuracy", cv=kfold))
```

```python
accuracy_results1
```

```python
accuracy_means1 = []
for e in accuracy_results1:
    accuracy_means1.append(e.mean()*100)
```

```python
accuracy_means1
```

```python
eval_frame['kfold_5']=accuracy_means1
eval_frame
```

# Stratifiedkfold_5

#### Stratified K-Folds cross-validator.It provides train/test indices to split data in train/test sets.This cross-validation object is a variation of KFold that returns stratified folds. The folds are made by preserving the percentage of samples for each class

```python
from sklearn.model_selection import StratifiedKFold
Stratifiedkfold = StratifiedKFold(n_splits=5)
# Modeling step Test differents algorithm
classifiers_4 = []
classifiers_4.append(KNeighborsClassifier())
classifiers_4.append(LogisticRegression())
classifiers_4.append(GaussianNB())
classifiers_4.append(RandomForestClassifier())

accuracy_results_4 = []
for a in classifiers_4:
    accuracy_results_4.append(cross_val_score(a, train_input_SMOTE, train_output_SMOTE, scoring= "accuracy", cv=Stratifiedkfold))
accuracy_means_4 = []
for e in accuracy_results_4:
    accuracy_means_4.append(e.mean()*100)
accuracy_means_4
eval_frame['Stratifiedkfold_5']=accuracy_means_4

eval_frame
```
